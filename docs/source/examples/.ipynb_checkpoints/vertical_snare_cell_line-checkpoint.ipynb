## Vertical integration for paired datasets

uniPort also involves a vertical integration method for paired-cell datasets. Here we use uniPort to integrates paired SNARE-seq CellLineMixture datasets from [GSE126074](https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE126074).

```python
import uniport as up
import numpy as np
import pandas as pd
import scanpy as sc
import episcanpy as epi
```

```python
labels = pd.read_csv('snare/cell_line_meta.txt', sep='\t')
celltype = labels['cell_line'].values
```

```python
adata_atac = up.load_file('snare/GSE126074_CellLineMixture_SNAREseq_chromatin_counts.tsv')
adata_rna = up.load_file('snare/GSE126074_CellLineMixture_SNAREseq_cDNA_counts.tsv')
print(adata_atac)
print(adata_rna)
```

```python
adata_atac.obs['cell_type'] = celltype
adata_atac.obs['domain_id'] = 0
adata_atac.obs['domain_id'] = adata_atac.obs['domain_id'].astype('category')
adata_atac.obs['source'] = 'ATAC'

adata_rna.obs['cell_type'] = celltype
adata_rna.obs['domain_id'] = 1
adata_rna.obs['domain_id'] = adata_rna.obs['domain_id'].astype('category')
adata_rna.obs['source'] = 'RNA'
```

Preprocess scATAC-seq peaks. Select 2,000 highly variable peaks. Perform UMAP visualization.

```python
adata_atac.X[adata_atac.X>1] = 1
epi.pp.select_var_feature(adata_atac, nb_features=2000, show=False, copy=False)
sc.pp.normalize_total(adata_atac)
up.batch_scale(adata_atac)
print(adata_atac)

adata_atac_copy = adata_atac.copy()
sc.pp.pca(adata_atac_copy)
sc.pp.neighbors(adata_atac_copy)
sc.tl.umap(adata_atac_copy, min_dist=0.1)
sc.pl.umap(adata_atac_copy, color=['source', 'cell_type'])
```

Preprocess scRNA-seq peaks. Select 2,000 highly variable genes. Perform UMAP visualization.

```python
sc.pp.normalize_total(adata_rna)
sc.pp.log1p(adata_rna)
sc.pp.highly_variable_genes(adata_rna, n_top_genes=2000, inplace=False, subset=True)
up.batch_scale(adata_rna)
print(adata_rna)

adata_rna_copy = adata_rna.copy()
sc.pp.pca(adata_rna_copy)
sc.pp.neighbors(adata_rna_copy)
sc.tl.umap(adata_rna_copy, min_dist=0.1)
sc.pl.umap(adata_rna_copy, color=['source', 'cell_type'])
```

Project scATAC-seq into latent space with the help of scRNA-seq. The latent representations of scATAC-seq are stored at `adata_atac.obs['latent']`

```python
adata_atac = up.Run(adatas=[adata_atac, adata_rna], mode='v', lr=0.001, iteration=10000)
```

Perform UMAP visualization for scATAC latent representations after vertical integration. The cell types of scATAC-seq now are more distinguishable than before. 

```python
sc.pp.neighbors(adata_atac, use_rep='latent')
sc.tl.umap(adata_atac, min_dist=0.1)
sc.pl.umap(adata_atac, color=['source', 'cell_type'])
```
